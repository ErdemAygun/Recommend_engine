```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = ["Julia loves John more than Betty", "John loves Betty more than Julia", "Julia and Betty love John", "Today is raining"]
```

```python
cv = CountVectorizer()
wm =cv.fit_transform(sentences)
```

```python
pd.DataFrame( wm.toarray(), columns=cv.get_feature_names_out())
```

```python
pd.DataFrame(cosine_similarity(wm))
```

```python
df = pd.read_csv("http://51.91.120.89/ML0901/movie_dataset.csv")
df
```

```python
df.columns
```

```python
features = ['genres', 'keywords', 'cast', 'title', 'director']
```

```python
df.isna().sum()
```

```python
for f in features:
    df[f].fillna("", inplace=True)
```

```python
df["recommend"] = df.genres+" "+df.keywords+" "+df.cast+" "+df.title+" "+df.director
```

```python
df["recommend"].iloc[4]
```

```python
cv = CountVectorizer()
wm = cv.fit_transform(df.recommend)
wm.shape
```

```python
cosine_sim = cosine_similarity(wm)
cosine_sim.shape
```

```python
list(cosine_sim[0])
```

```python
list(enumerate(cosine_sim[0]))
```

```python
sorted(list(enumerate(cosine_sim[0])), reverse=True)
```

```python
sorted(list(enumerate(cosine_sim[0])), reverse=True, key=lambda x:x[1])
```

```python
def get_title_from_index(index):
    return df[df.index==index].title.values[0]
```
